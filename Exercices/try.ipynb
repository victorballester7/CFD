```python
import numpy as np
```

```python
# print all array in one line
np.set_printoptions(linewidth=1000)
```

```python
# explicit euler method for heat equation

# parameters
L = 1.0
Nx = 10
Nt = 100
kappa = 1.0
dx = L / Nx
dt = 0.01/2+0.001
lamb = kappa * dt / dx**2
print(lamb)

# initial condition
u = np.zeros(Nx)
# change only the middle point to 1
u[Nx//2] = 1.0
print(u)

U = np.zeros((Nt, Nx))
U[0] = u
# time stepping
for n in range(Nt):
    unew = np.zeros(Nx-2)
    unew = (1 - 2*lamb) * u[1:-1] + lamb * (u[:-2] + u[2:])
    u[1:-1] = unew
    U[n] = u
print(U)
```
